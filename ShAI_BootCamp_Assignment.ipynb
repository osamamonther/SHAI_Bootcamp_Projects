# 3)Data Cleaning

There are some negative values in some columns, such as the TotalPay column, but I will not modify these values.
```python
# Show the rows from salary column which have negative value
df[df["TotalPay"] < 0]
```

```python
# Check for the null values
df.isnull().sum()
```

```python
# Delete Notes column and Status column, because all value is null
df.drop("Notes", axis=1, inplace=True)
df.drop("Status", axis=1, inplace=True)
df.shape
```

```python
# Show the rows from OtherPay column which have null values
df[df["OtherPay"].isnull()]
```

```python
# Delete the null values from OvertimePay column and OtherPay column
df.dropna(subset=["OvertimePay", "OtherPay"], inplace=True)
df[["OvertimePay", "OtherPay"]].isnull().sum()
```

```python
# Show the rows from BasePay column which have null values
df[df["BasePay"].isnull()]
```

```python
# Replace the null value with zero from BasePay column
df.fillna({"BasePay": 0}, inplace=True)
df[["BasePay"]].isnull().sum()
```

```python
# Show the rows from Benefits column which have null values
df_0 = df[df["Benefits"].isnull()]
df_0.sample(5)
```

```python
# Replace the null value with zero from Benefits column
df.fillna({"Benefits": 0}, inplace=True)
df[["Benefits"]].isnull().sum()
```

```python
# Now the Data don't have null values
df.isnull().sum()
```

# 4)Basic Data Visualization

```python
# Show histogram (distribution) for Salary Column
sns.histplot(data = df, x ='TotalPay')
```

```python
# Show the Pie chart for JobTitle Column
plt.pie(df["JobTitle"].value_counts(), labels= df["JobTitle"].value_counts().index,
        wedgeprops={'edgecolor' : 'white'},
        autopct = '%1.f%%')
plt.title('Pie Chart for employees departments')
plt.show()
```

# 5)Grouped Analysis

```python
# Show the descriptive statistics of salary for each year
df.groupby("Year")[["TotalPay"]].describe()
```

```python
# Show the mean salary for each year and sort by the highest mean
df.groupby("Year")[["TotalPay"]].mean().sort_values(by = "TotalPay", ascending = False)
```

```python
# Show the descriptive statistics of salary for each Job Title
df.groupby("JobTitle")[["TotalPay"]].describe()#.sort_values(by = "count", ascending = False)
```

```python
# Show the mean salary for each Job Title and sort by the highest mean
df.groupby("JobTitle")[["TotalPay"]].mean().sort_values(by = "TotalPay",
                                                        ascending = False)
```

```python
# Show the count of Employee for each Job Title
df.groupby("JobTitle")[["Id"]].count().sort_values(by = "Id", ascending = False)
```

# 6)Simple Correlation Analysis

```python
# Show the correlation between OtherPay Column and Salary Column
df[["OtherPay","TotalPay"]].corr()
```

```python
# Show the correlation between numerical columns
df_numerical = df.select_dtypes(include=['int','float'])

df_numerical.drop("Id", axis=1, inplace=True)
df_numerical.drop("Year", axis=1, inplace=True)

df_numerical.corr()
```

```python
# show the Heat Map
sns.heatmap(df_numerical.corr(), annot = True,
            cmap = 'coolwarm', fmt = '.2f',
            vmin= -1, vmax = 1)
plt.show()
```

```python
# Show the relationship between OvertimePay column and salary column by Scatter Plot
sns.scatterplot(data= df, x= 'OvertimePay', y= 'TotalPay')
plt.show()
```

```python
# Show the relationship between Benefits column and salary column by Scatter Plot
sns.scatterplot(data= df, x= 'Benefits', y= 'TotalPay')
plt.show()
```
